```python
col1 = list(dfTitanic.columns.values)
col2 = [str(dfTitanic[str(col1[a])].dtype) for a in range(len(col1))]
col3 = [len(dfTitanic[dfTitanic[str(col1[b])].isna() == True]) for b in range(len(col1))]
col4 = list(np.round([(int(col3[c]) / 891) * 100 for c in range(len(col1))], 2))
col5 = [dfTitanic[str(col1[d])].nunique() for d in range(len(col1))]
col6 = [list(dfTitanic[str(col1[e])].sample(5).unique()) for e in range(len(col1))]

dfTable = pd.DataFrame({
    "dataFeatures": col1,
    "dataType": col2,
    "null": col3,
    "nullPct": col4,
    "unique": col5,
    "uniqueSample": col6
})

dfTable
```

### 9. Adding New Features/Columns/Variables

#### Buat DataFrame baru dengan column2 sebagai berikut
 - __id__ = id table ini (PK)
 - __mean_fare__ = Mean Fare
 - __std_fare__ = Std Fare
 - __batas_min_fare__ = Mean Fare - (3 x Std Fare)
 - __batas_max_fare__ = Mean Fare + (3 x Std Fare)
 
#### Dan data2nya ada 3 (fare setiap class yaitu first, second, and third class)

```python
col1 = [1, 2, 3]
classUnique = dfTitanic["class"].unique()
col2 = [dfTitanic[dfTitanic["class"] == str(a)]["fare"].mean() for a in classUnique]
col3 = [dfTitanic[dfTitanic["class"] == str(b)]["fare"].std() for b in classUnique]
col4 = [col2[c] - (3 * col3[c]) for c in range(len(col1))]
col5 = [col2[d] + (3 * col3[d]) for d in range(len(col1))]

dfTable2 = pd.DataFrame({
    "id": col1,
    "mean_fare": col2,
    "std_fare": col3,
    "batas_min_fare": col4,
    "batas_max_fare": col5
})
dfTable2
```

#### Print outlier % of each class

```python
outThird = (len(dfTitanic[dfTitanic["class"] == "Third"]["fare"] < dfTable2["batas_min_fare"][0]) + len(dfTitanic[dfTitanic["class"] == "Third"]["fare"] > dfTable2["batas_max_fare"][0])) / len(dfTitanic[dfTitanic["class"] == "Third"]) * 100
outThird
```
